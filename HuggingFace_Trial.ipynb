#Calling Hugging Face API
### Hugging face endpoints

```python
from langchain_huggingface import HuggingFaceEndpoint
import os

os.environ["HUGGINGFACEHUB_TOKEN"] = sec_key
```

```python
repoid="mistralai/Mistral-7B-Instruct-v0.3"
llm=HuggingFaceEndpoint(repo_id=repoid,temperature=0.7,provider="hf-inference")
```

```python
llm.invoke("what is the capital of india")
```

```python
from re import template
from langchain import PromptTemplate, LLMChain

ques="Who won 2011 cricket world cup?"

template="""Question: {question}
Answer: Let's think step by step."""

prompt=PromptTemplate(template=template, input_variables=["question"])

print(prompt)
```

```python
llm_chain=LLMChain(prompt=prompt, llm=llm)
print(llm_chain.invoke(ques))
```

#Hugging Face Pipeline

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline, AutoModelForCausalLM


model="gpt2"
ai_model=AutoModelForCausalLM.from_pretrained(model)
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
pipe=pipeline("text-generation", model=ai_model,tokenizer=tokenizer)
hf=HuggingFacePipeline(pipeline=pipe)
```

```python
hf.invoke("Langchain is a company")
```
